```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
import seaborn as sns
```

```python
energyresults = pd.read_csv('/content/HRCE_res.csv')
incomeresults = pd.read_csv('/content/finOccupied.csv')
```

```python
energydf = pd.DataFrame(energyresults)
incomedf = pd.DataFrame(incomeresults)

```

```python
energydf.info()
```

```python
energy_subset = energydf[['county_name', 'population', 'occuped housing units',
                          'Electricity GHG emissions mtons CO2e','Natural Gas GHG emissions mtons CO2e']]
energy_subset
```

```python
incomedf
```

```python
incomedf.info()
```

```python
income_subset = incomedf.iloc[:, [0, 1, 14, 26]]
```

```python
# Reset the index and drop the old index
income_subset.reset_index(drop=True, inplace=True)

# Change column names
income_subset.columns = ['Virginia County', 'Occupied housing units','Median household income (dollars)','Median Monthly Housing Costs (dollars)']

# Drop a row by index
income_subset.drop(0, inplace=True)
income_subset.drop(9, inplace=True)
income_subset.reset_index(drop=True, inplace=True)

income_subset
```

```python
energy_subset
```

```python
# Select specific rows by index position
filtered_energy_subset = energy_subset.iloc[[0,1,2,3,4,7,10,11]]
filtered_energy_subset.reset_index(drop=True, inplace=True)
filtered_energy_subset
```

```python
combined_df = pd.concat([filtered_energy_subset,income_subset], axis=1)
combined_df
```

```python
combined_df.info()
```

```python
# Remove commas from the specified columns and convert them to integers
combined_df['Occupied housing units'] = combined_df['Occupied housing units'].replace({',': ''}, regex=True).astype(int)
combined_df['Median household income (dollars)'] = combined_df['Median household income (dollars)'].replace({',': ''}, regex=True).astype(int)
combined_df['Median Monthly Housing Costs (dollars)'] = combined_df['Median Monthly Housing Costs (dollars)'].replace({',': ''}, regex=True).astype(int)

combined_df
```

```python
combined_df['Occupied Housing Units Averaged'] = (combined_df['occuped housing units'] + combined_df['Occupied housing units']) / 2
combined_df
```

```python
# Convert median household income to monthly income
combined_df['Monthly Household Income'] = combined_df['Median household income (dollars)'] / 12

# Calculate the proportion of housing costs to monthly income
combined_df['Housing Costs to Income Proportion'] = combined_df['Median Monthly Housing Costs (dollars)'] / combined_df['Monthly Household Income']

combined_df
```

```python
combined_df['Electricity Emissions per Housing Unit'] = combined_df['Electricity GHG emissions mtons CO2e'] / combined_df['Occupied Housing Units Averaged']

combined_df['Natural Gas Emissions per Housing Unit'] = combined_df['Natural Gas GHG emissions mtons CO2e'] / combined_df['Occupied Housing Units Averaged']

combined_df
```

```python
combined_df['county_name'] = combined_df['county_name'].str.replace('city', '', case=False)

```

```python
combined_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'combined_df' is the DataFrame containing the data

# Set up the plot
plt.figure(figsize=(12, 8))

# Scatter plot for Electricity (using circles 'o')
sns.scatterplot(data=combined_df,
                x='Electricity Emissions per Housing Unit',
                y='Median household income (dollars)',
                marker='o',
                color='blue',
                label='Electricity',
                s=100)  # Adjust size for better visibility

# Scatter plot for Natural Gas (using squares 's')
sns.scatterplot(data=combined_df,
                x='Natural Gas Emissions per Housing Unit',
                y='Median household income (dollars)',
                marker='s',
                color='red',
                label='Natural Gas',
                s=100)  # Adjust size for better visibility

# Add labels for each point (both Electricity and Natural Gas)
for i in range(len(combined_df)):
    # For electricity points (blue circles)
    plt.text(combined_df['Electricity Emissions per Housing Unit'][i] + 0.05,  # X position (adjust for spacing)
             combined_df['Median household income (dollars)'][i],  # Y position for electricity (income)
             combined_df['county_name'][i],  # Label text (county name)
             fontsize=9,  # Font size
             ha='left',  # Horizontal alignment
             va='bottom',  # Vertical alignment
             color='black')  # Label color

    # For natural gas points (red squares)
    plt.text(combined_df['Natural Gas Emissions per Housing Unit'][i] + 0.05,  # X position (adjust for spacing)
             combined_df['Median household income (dollars)'][i],  # Y position for natural gas (income)
             combined_df['county_name'][i],  # Label text (county name)
             fontsize=9,  # Font size
             ha='left',  # Horizontal alignment
             va='bottom',  # Vertical alignment
             color='black')  # Label color

# Add labels and title
plt.xlabel('GHG Emissions per Housing Unit (Metric Tons CO2e)')
plt.ylabel('Median Household Income (Dollars)')
plt.title('GHG Emissions per Housing Unit vs Median Household Income')

# Set the y-axis to start at 0
plt.ylim(bottom=0)

# Display the legend
plt.legend(title='Energy Source')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn.objects as so
import matplotlib.pyplot as plt
import numpy as np

# Transform the data into a long format
melted_df = pd.melt(combined_df,
                    id_vars=['county_name', 'Median household income (dollars)'],
                    value_vars=['Electricity Emissions per Housing Unit', 'Natural Gas Emissions per Housing Unit'],
                    var_name='Energy Source',
                    value_name='GHG Emissions per Housing Unit')

melted_df['Energy Source'] = melted_df['Energy Source'].replace({
    'Electricity Emissions per Housing Unit': 'Electricity',
    'Natural Gas Emissions per Housing Unit': 'Natural Gas'
})

text = pd.DataFrame({
    'x': [20000],
    'y': [1.5],
    'text': ["Natural Gas"]
    })
text.head()

text2 = pd.DataFrame({
    'x2': [20000],
    'y2': [5],
    'text2': ["Electricity"]
    })
text2.head()

text3 = pd.DataFrame({
    'x3': [50000],
    'y3': [3],
    'text3': ["Lowest Income Counties"]
    })
text3.head()

text4 = pd.DataFrame({
    'x4': [85000],
    'y4': [3.5],
    'text4': ["Highest Income Counties"]
    })
text4.head()

text5 = pd.DataFrame({
    'x5': [70000],
    'y5': [6.5],
    'text5': ["Notice these counties have higher \nemissions but lower incomes"]
    })
text5.head()



# Transform the data into a long format
melted_df = pd.melt(combined_df,
                    id_vars=['county_name', 'Median household income (dollars)'],
                    value_vars=['Electricity Emissions per Housing Unit', 'Natural Gas Emissions per Housing Unit'],
                    var_name='Energy Source',
                    value_name='GHG Emissions per Housing Unit')

# Replace the long format values to represent the Energy Source
melted_df['Energy Source'] = melted_df['Energy Source'].replace({
    'Electricity Emissions per Housing Unit': 'Electricity',
    'Natural Gas Emissions per Housing Unit': 'Natural Gas'
})

# Sort the data first by emissions type and then by median income within each type
melted_df = melted_df.sort_values(by=['Energy Source', 'Median household income (dollars)']).reset_index(drop=True)

# Initialize the 'label_position' column with the GHG emissions value
melted_df['label_position'] = melted_df['GHG Emissions per Housing Unit']

# Apply alternating positions based on the sorted order
for source in melted_df['Energy Source'].unique():
    # Select rows for the current energy source
    source_data = melted_df[melted_df['Energy Source'] == source]

    # Apply alternating positions within the emissions type group
    source_data.loc[source_data.index % 2 == 0, 'label_position'] -= 0.3  # Even indices shift up
    source_data.loc[source_data.index % 2 != 0, 'label_position'] += 0.3  # Odd indices shift down

    # Update the main DataFrame with the modified positions
    melted_df.loc[source_data.index, 'label_position'] = source_data['label_position']

# variables for title and subtitle
title = "Hampton Roads Energy Consumption and Income Rates"
subtitle = "Higher Residential GHG Emissions are not directly correlated to Higher Income in the Hampton Roads"

# Now plot using melted_df
p = (
    so.Plot(melted_df, x='Median household income (dollars)', y='GHG Emissions per Housing Unit',
            color='Energy Source')  # Map color to emission type
    .layout(size=(10, 8))
    .add(so.Dot(pointsize=10))  # Increase marker size
    .theme({**axes_style("whitegrid")})  # Change theme
    .limit(x=(0, 100000), y=(0, 8))  # Adjust axis limits
    .label(title=title + "\n" + subtitle)
    .label(x="Median Household Income (Dollars)", y="GHG Emissions per Housing Unit (Metric Tons CO2e)")
    .label(legend=False)  # Remove legend if not needed
    # Add text labels for counties with alternating positions by emissions type
    .add(so.Text(color="k", fontsize=10),
         data=melted_df, x='Median household income (dollars)', y='label_position', text='county_name')
    # Add text labels for emissions types
    .add(so.Text(color="k", fontsize=15),
         data=text, x='x', y='y', text='text')
    .add(so.Text(color="k", fontsize=15),
         data=text2, x='x2', y='y2', text='text2')
    .add(so.Text({"fontweight": "bold"}, color="k", fontsize=10),
         data=text5, x='x5', y='y5', text='text5')
)

p.theme({"axes.facecolor": "w"})

```
